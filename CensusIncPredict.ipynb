# United-States-Census Income Prediction
Build a classification model based on the features that you select to predict person's yearly income is above $50k or not.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

The data here is for the "Census Income" dataset, which contains data on adults from the 1994 census. This data is labeled with whether the person's yearly income is above or below $50K (and you are trying to model and predict this).

```python
cdata=pd.read_csv("census - census.csv")
cdata.head()
```

### Performing Data Cleaning on the dataset

Now we have to handle some Miscoded values-->'?' in our dataset

```python
cdata=cdata.replace('?',np.nan)
```

```python
cdata
```

```python
cdata.describe()
```

```python
cdata.mode()
```

```python
cdata.isnull().sum() #summing up for all null values in dataset
```

Handling missing data with their "modes" of respective coloumn cause of categorical data.

```python
cdata["workclass"]=cdata["workclass"].fillna("Private")
cdata["occupation"]=cdata["occupation"].fillna("Prof-specialty")
cdata["native.country"]=cdata["native.country"].fillna("United-States")
```

```python
# converting sex and income to "0" and "1" 
def income_to_numeric(x):
    if x=='>50K':
        return 1
    if x=='<=50K':
        return 0
    
cdata['income'] = cdata['income'].apply(income_to_numeric)
def sex_to_numeric(x):
    if x=='Male':
        return 1
    if x=='Female':
        return 0
    
cdata['sex'] = cdata['sex'].apply(sex_to_numeric)
```

```python
dumydata=pd.get_dummies(cdata, columns=["workclass", "education", "marital.status", "occupation"])
dumydata
```

### Feature Engineering
Extracting Out important features from dataset

```python
dumydata=dumydata.drop(columns=["relationship","race","native.country"])
```

After performing EDA we came to a conclusion that there are three columns which are not useful for the prediction and removal of these columns will not affect the prediction

We also converted required Categorical values into Numerical values so that it will be easy to correlate between them.

```python
X = dumydata.drop(['income'],axis=1)    
Y = dumydata['income']
```

```python
X = (X-X.min())/(X.max()-X.min())
X.head()
```

```python
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(0.005)
X = selector.fit_transform(X)
X.shape,Y.shape
```

### PCA

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_Std=StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
```

```python
principalComponents = pca.fit_transform(x_Std)
```

```python
principalDf = pd.DataFrame(data = principalComponents)
```

```python
principalDf
```

```python
from sklearn.model_selection import train_test_split
principalDf_train,principalDf_test,Y_train,Y_test = train_test_split(principalDf,Y,test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(principalDf_train,Y_train)
```

```python
pred=model.predict(principalDf_test)
from sklearn import metrics
pca_acc=metrics.accuracy_score(Y_test,pred)*100
pca_acc
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
reg_lr = LogisticRegression(random_state=5)
reg_lr.fit(principalDf_train,Y_train)
```

```python
pred_lr=reg_lr.predict(principalDf_test)
from sklearn import metrics
lr_acc=metrics.accuracy_score(Y_test,pred_lr)*100
lr_acc
```

### KNN

### VotingClassifier

```python
from sklearn.ensemble import VotingClassifier
model1 = DecisionTreeClassifier()
model2 = LogisticRegression()
model3 = SVC()
model_E_VC = VotingClassifier(estimators=[('DT',model1),('LR',model2),('SVC',model3)])
model_E_VC.fit(principalDf_train,Y_train)
pred_E_VC = model_E_VC.predict(principalDf_test)
vc=metrics.accuracy_score(Y_test,pred_E_VC)*100
vc
```
